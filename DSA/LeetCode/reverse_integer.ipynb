7. Reverse Integer

Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

```python
def reverse(x):
        """
        :type x: int
        :rtype: int
        """
        sign = -1 if x < 0 else 1
        abs_n = abs(x)
        temp = 0

        while abs_n > 0:
            digit = abs_n % 10
            temp = (temp * 10) + digit
            abs_n //= 10

        return temp * sign


print(reverse(-123))
```
