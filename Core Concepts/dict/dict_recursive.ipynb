Find out the answer

input -> a = {"a":{"b":{"c":1}},"d":2}
output -> {"a.b.c" : 1, "d":2}


```python
# 1. recursive method
class Dict():

    def __init__(self):
        pass

    def recursive(self,obj,update_key= None):
        result = {}
        for key,value in obj.items():
            modify_key = f"{update_key}.{key}" if update_key else key
            if isinstance(value,dict):
                result.update(self.recursive(value,modify_key))
            else:
                result[modify_key] = value
        return result


if __name__ == "__main__":
    dictionary = Dict()
    input = {"a":{"b":1},"d":2}
    print(dictionary.recursive(input))
```
