````python
def init_only(func):
    print( "deco received",func.__name__)
    def wrapper(*args):
        if not all(isinstance(i,int) for i in args):
            raise TypeError("All Arguements must be integers")
        return func(*args)
    return wrapper


@init_only
def add(a,b):
    return a+b

print(add(5,4))
````
