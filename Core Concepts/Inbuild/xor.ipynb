
## ✅ **Binary Place Values**

```
2⁰ = 1  
2¹ = 2  
2² = 4  
2³ = 8  
2⁴ = 16  
2⁵ = 32  
2⁶ = 64  
2⁷ = 128  
2⁸ = 256  
2⁹ = 512  
...
```

This is why binary digits (bits) represent these values in each position.

## Example

Binary number **1011** means:

```
1 × 2³  = 8
0 × 2²  = 0
1 × 2¹  = 2
1 × 2⁰  = 1
----------------
Total = 11
```
