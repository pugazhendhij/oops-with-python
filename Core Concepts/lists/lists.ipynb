### Python List concepts

ðŸ“Œ 1. sorted()

Function (built-in)

Works on any iterable (list, tuple, set, dict keys/values, string)

Returns a new list â†’ original data is not modified

Can be used in expressions or assigned to variables

```python
arr = ["banana","cherry","mango","apple"]
result = sorted(arr)
print(f"Orginal List : {arr} \nSorted List : {result}")
```

ðŸ“Œ 2. .sort()

List method only â†’ works only on lists

Sorts the list in-place â†’ modifies the original list

Returns None (so you cannot assign it)

```python
arr = ["banana","cherry","mango","apple"]
arr.sort()


print(f"Orginal List : {arr}")
```

shuffle the list without using default shuffle


```python
import random

# 1. inbuild method
arr = [1,4,2,3,7,8,9,10,2,13,42,45]
result = random.shuffle(arr)

# 2. using for loop
'''
- randint(a, b) includes both ends â†’ a â‰¤ n â‰¤ b  may be return length it throws error list index out of range 
- randrange(start, stop) never includes stop â†’ start â‰¤ n < stop it works correctly
'''
def shuffle_list_using_for(arr):
    for i in range(len(arr)):
        # index = random.randint(0,len(arr))
        index = random.randrange(0,len(arr))
        temp = arr[i]
        arr[i] = arr[index]
        arr[index] = temp
    return arr

# 3. using while loop
'''
- here we use the temp arr and remove the index one by one with range of len(temp) because it should remove every time
'''
def shuffle_list_using_while(arr):
    result_list = [] 
    temp = arr
    while temp:
        index = random.randrange(0,len(temp))
        result_list.append(temp.pop(index))

    return result_list

if __name__ == "__main__":
    print(shuffle_list_using_for(arr))
    print(shuffle_list_using_while(arr))
```
