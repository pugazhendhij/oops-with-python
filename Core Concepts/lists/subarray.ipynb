Given an integer array arr and an integer k, find the maximum sum of any contiguous subarray of size k.

If the array is empty, return 0.

1. 
    Input:
    arr = [1, 2, 1, 5, 1, 3, 4]
    k = 3

    Output: 9 - > [5,1,3]

2. 
    Input:
    arr = [2, 3, 1, 4, 3]
    k = 2

    Output: 7 -> [4,3]

```python
class Subarray():

    def __init__(self,arr):
        self.arr = arr

    
    def find_subarray(self,k,target):
        for i in range(len(self.arr) - k + 1):
            window = self.arr[i: i+k]
            if sum(window) == target:
                return window
        return 0

if __name__ == "__main__":
    arr = Subarray([1, 2, 1, 5, 1, 3, 4])
    print(arr.find_subarray(3,9))

    arr1 = Subarray([2, 3, 1, 4, 3])
    print(arr1.find_subarray(2,7))
```
