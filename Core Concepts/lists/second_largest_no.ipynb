````python
def find_second_largest(arr):
    first = second = float('-inf')
    
    for num in arr:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
        print(f"num: {num} || first {first} || second {second}")
    return second


data = [1, 2, 3, 4, 6, 9, 10, 12, 25, 10, 3, 4, 25]
print(find_second_largest(data))
````
