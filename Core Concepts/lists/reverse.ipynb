````python
import time

class ReverseList:

    def list_reverse_for(self, data):
        result = []
        for i in range(len(data)-1, -1, -1):
            result.append(data[i])
        return result

    def list_reverse_while(self, data):
        result = []
        i = len(data) - 1
        while i >= 0:
            result.append(data[i])
            i -= 1
        return result

    def complexity(self, function, data):
        start = time.time()
        value = function(data)
        end = time.time()
        print(f"{function.__name__} â†’ Time taken: {end - start} seconds")
        return value

if __name__ == "__main__":
    data = [i for i in range(10)]
    rev = ReverseList()
    rev.complexity(rev.list_reverse_for, [i for i in range(1000)])
    rev.complexity(rev.list_reverse_while, [i for i in range(1000)])
````

````python
def two_pointers_reverse(arr):
    left = 0
    right = len(arr) - 1


    while left < right:
        arr[left],arr[right] = arr[right],arr[left]

        left += 1
        right -= 1

    return arr

data = [i for i in range(9)]
print(two_pointers_reverse(data))
````
